```python
import matplotlib.pyplot as plt

# Assuming you have calculated the average accuracy and loss and loaded the data for w1, w2, w3, w4, w5, w6

# Calculate average accuracy and loss across workers for each epoch
avg_accuracy = (w1['Accuracy'] + w2['Accuracy'] + w3['Accuracy'] + w4['Accuracy'] + w5['Accuracy'] + w6['Accuracy']) / 6
avg_loss = (w1['Loss'] + w2['Loss'] + w3['Loss'] + w4['Loss'] + w5['Loss'] + w6['Loss']) / 6

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(w1['Epoch'], avg_accuracy, label='Average Accuracy')
plt.plot(w1['Epoch'], avg_loss, label='Average Loss')
plt.xlabel('Epoch', fontsize=14, fontweight='bold')
plt.ylabel('Value', fontsize=14, fontweight='bold')
plt.title('Scalability Analysis', fontsize=16, fontweight='bold')
plt.legend(fontsize=12, loc='center right').set_title('Metrics', prop={'weight': 'bold'})  # Bold legend title and labels
plt.grid()
plt.show()
```

```python
# Calculate standard deviations of accuracy and loss for each worker
std_accuracy = np.std([w1['Accuracy'], w2['Accuracy'], w3['Accuracy'], w4['Accuracy'], w5['Accuracy'], w6['Accuracy']], axis=0)
std_loss = np.std([w1['Loss'], w2['Loss'], w3['Loss'], w4['Loss'], w5['Loss'], w6['Loss']], axis=0)

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot accuracy std dev on the left y-axis
ax1.plot(w1['Epoch'], std_accuracy, label='Accuracy Std Dev', color='tab:blue')
ax1.set_xlabel('Epoch', fontsize=14, fontweight='bold')
ax1.set_ylabel('Accuracy Std Dev', fontsize=14, fontweight='bold', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a twin y-axis on the right for loss std dev
ax2 = ax1.twinx()
ax2.plot(w1['Epoch'], std_loss, label='Loss Std Dev', color='tab:orange')
ax2.set_ylabel('Loss Std Dev', fontsize=14, fontweight='bold', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Title and legend
plt.title('Reliability Analysis', fontsize=16, fontweight='bold')
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right', fontsize=12).set_title('Metrics', prop={'weight': 'bold'})

plt.grid()
plt.show()
```
